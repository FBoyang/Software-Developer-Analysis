```python
import numpy as np
import pandas as pd
df = pd.read_csv("FCC2.csv")
#let's drop the columns with NA > 3000
for i in range((df.shape)[1]):
    column = df.iloc[:, i]
    if(column.isnull().sum() <= 3000):
        if(i == 0):
            ndf = pd.DataFrame(data = column)
        else:
            ndf = ndf.join(column)
ndf.to_csv("FCCn.csv" ,sep = ',')
```
